## 2 - Recherche des clusters par la méthodes des K-moyennes

**À l'aide des techniques vues en cours, recherchez les clusters présents dans le jeu de données. Trouver le nombre de clusters K présents**
<br/>**À vous de jouer !**

```python
from sklearn.cluster import KMeans

disto = []
for k in range(1,10):
    kmeans = KMeans(n_clusters=k, n_init=100).fit(X)
    disto.append(kmeans.inertia_)
```

```python
plt.plot(disto)
```
